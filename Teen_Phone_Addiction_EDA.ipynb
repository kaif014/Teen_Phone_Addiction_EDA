## Visual Exploration of Key Relationships

In this section, we visualize important relationships and patterns within the dataset using various types of plots. These graphs help us better understand how different factors such as screen time, sleep hours, academic performance, and phone usage behaviors are interconnected.

### Scatterplot: Daily Usage Hours vs Academic Performance

This scatterplot shows how the number of daily phone usage hours correlates with academic performance, colored by gender.

```python
sns.scatterplot(data=df_addiction, x="Daily_Usage_Hours", y="Academic_Performance", hue="Gender")
plt.title("Scatterplot: Daily Usage Hours vs Academic Performance")
plt.show()
```

### Lineplot: Average Sleep Hours by Age

This lineplot displays how average sleep hours vary across different age groups.

```python
avg_sleep_by_age = df_addiction.groupby('Age', as_index=False)['Sleep_Hours'].mean()

sns.lineplot(data=avg_sleep_by_age, x='Age', y='Sleep_Hours', marker='o')
plt.title("Average Sleep Hours by Age")
plt.xlabel("Age")
plt.ylabel("Average Sleep Hours")
plt.show()
```

### Barplot: Average Addiction Level by Phone Usage Purpose

This barplot shows the average addiction levels based on different primary purposes of phone usage.

```python
sns.barplot(data=df_addiction, x="Phone_Usage_Purpose", y="Addiction_Level", estimator='mean', errorbar=None)
plt.title("Barplot: Average Addiction Level by Phone Usage Purpose")
plt.xticks(rotation=45)
plt.show()
```

### Boxplot: Screen Time Before Bed by Gender

This boxplot highlights the distribution of screen time before bed across different genders.

```python
sns.boxplot(data=df_addiction, x="Gender", y="Screen_Time_Before_Bed")
plt.title("Boxplot: Screen Time Before Bed by Gender")
plt.show()
```

### Histplot: Distribution of Phone Checks Per Day

This histogram displays the distribution of how frequently teens check their phones each day.

```python
sns.histplot(data=df_addiction, x="Phone_Checks_Per_Day", bins=16, kde=True)
plt.title("Histplot: Distribution of Phone Checks Per Day")
plt.show()
```

### Correlation Analysis

To understand the relationships between numerical variables, we generate a correlation heatmap. This allows us to identify strong positive or negative correlations that might reveal behavioral or psychological patterns related to phone addiction among teens.

```python
corr = df_addiction.select_dtypes(include=['number']).corr()

plt.figure(figsize=(12, 8))
ax = sns.heatmap(
    corr.values,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1, vmax=1,
    linewidths=0.5,
    linecolor="white",
    square=False,
    annot_kws={"ha": "center", "va": "center"},
    xticklabels=corr.columns,
    yticklabels=corr.index
)
plt.title("Correlation Heatmap", fontsize=16, pad=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Heatmap Insights

From the correlation heatmap:

- **Addiction_Level** shows a strong positive correlation with:
  - **Daily_Usage_Hours (0.60)**: Suggesting that increased daily phone use is highly associated with addiction.
  - **Phone_Checks_Per_Day (0.32)**, **Apps_Used_Daily (0.31)**, and **Screen_Time_Before_Bed (0.31)** also show moderate correlations, indicating consistent usage habits relate to higher addiction levels.

- **Time_on_Gaming** has a slight negative correlation with **Academic_Performance (-0.22)**, suggesting that excessive gaming might affect studies.

- **Time_on_Education** shows a mild positive correlation with **Academic_Performance (0.22)**.

This heatmap helps identify key behavioral patterns linked to addiction and academic outcomes.

### Regression Analysis of Behavioral Factors

We use regression plots to visualize and confirm linear trends between behavioral indicators and addiction level. These plots enhance the correlation analysis and provide clearer insights into how specific habits influence addiction.

#### Daily Usage Hours vs Addiction Level

This regression plot examines the linear relationship between daily screen time and addiction level. A strong upward trend suggests that more hours on the phone are closely linked with higher addiction scores.

```python
sns.regplot(data=df_addiction, x='Daily_Usage_Hours', y='Addiction_Level', scatter_kws={"alpha":0.5})
plt.title("Daily Usage Hours vs Addiction Level")
plt.xlabel("Daily Usage Hours")
plt.ylabel("Addiction Level")
plt.show()
```

#### Phone Checks Per Day vs Addiction Level

This plot shows how the frequency of phone checking throughout the day correlates with addiction levels. The positive slope suggests that frequent checking may indicate addictive behavior.

```python
sns.regplot(data=df_addiction, x='Phone_Checks_Per_Day', y='Addiction_Level', scatter_kws={"alpha":0.5})
plt.title("Phone Checks Per Day vs Addiction Level")
plt.xlabel("Phone Checks Per Day")
plt.ylabel("Addiction Level")
plt.show()
```

#### Apps Used Daily vs Addiction Level

This regression plot reveals the relationship between the number of apps used daily and addiction level. An upward trend reflects how broader app usage may be associated with higher phone dependence.

```python
sns.regplot(data=df_addiction, x='Apps_Used_Daily', y='Addiction_Level', scatter_kws={"alpha":0.5})
plt.title("Apps Used Daily vs Addiction Level")
plt.xlabel("Apps Used Daily")
plt.ylabel("Addiction Level")
plt.show()
```

# Conclusion

This exploratory data analysis provided meaningful insights into the patterns of teen smartphone usage and its relationship with mental well-being, academic performance, and behavioral traits.

### Key Observations:

### Screen Time & Addiction
• Daily screen time and addiction levels show a strong positive correlation, indicating that teens who spend more hours on their phones tend to exhibit higher signs of addiction.
• Phone checks per day, apps used daily, and screen time before bed are also moderately correlated with addiction, highlighting consistent usage habits as contributing factors.

### Mental Health Indicators
• Teens with higher phone usage report elevated levels of anxiety and depression, suggesting a potential mental health impact from excessive screen exposure.
• A negative trend was noticed between sleep hours and depression level, emphasizing the role of healthy sleep in emotional well-being.

### Academic Performance
• Time spent on gaming is slightly negatively correlated with academic performance, while time on educational apps shows a mild positive correlation.
• This suggests that how a teen uses their phone matters more than how long they use it.

### Usage Purpose & Behavior
• Teens primarily use their phones for browsing, education, and social media, followed closely by gaming.
• Despite varied purposes, behavioral indicators (like phone checks and app count) remain consistent among those with higher addiction scores.

### Correlation Insights
• The correlation heatmap helped identify behavioral patterns and psychological traits that cluster around high phone addiction.
• Features like self-esteem, family communication, and weekend usage provide context for deeper behavioral understanding.
